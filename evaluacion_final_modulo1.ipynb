2. ventas_totales(float): Un atributo para llevar un registro de las ventas totales de la tienda.Inicializado con valor 0.

· La clase TiendaOnline debe tener los siguientes métodos:

1. agregar_producto(self, nombre, precio, cantidad): Este método agrega un producto alinventario o actualiza su cantidad si ya existe. Debe recibir el nombre, precio y cantidad del productocomo parámetros.
      -Itera a través del inventario y compara los nombres de los productos con el nombre proporcionado.
      -Si el producto ya existe, actualiza la cantidad.
      -Si no existe, agrega un nuevo producto al inventario.

2. ver_inventario(self): Muestra el inventario de productos con sus detalles.
· Utiliza un bucle for para recorrer el inventario.
· Imprime los detalles (nombre, precio, cantidad) de cada producto.

3.  buscar_producto(self, nombre): Busca un producto en el inventario por nombre y muestra susdetalles si se encuentra. Debe recibir el nombre del producto como parámetro.
   - Utiliza un bucle for para recorrer el inventario.

4. actualizar_stock(self, nombre, cantidad): Actualiza el stock de un producto en elinventario. Debe recibir el nombre del producto y la cantidad a agregar o quitar como parámetros.
  - Utiliza un bucle for para recorrer el inventario.

  - Si el producto no esta en el inventario muestra un mensaje indicándolo.

5. eliminar_producto(self, nombre): Elimina un producto del inventario por nombre. Debe recibir el nombre del producto como parámetro.
  - Utiliza un bucle for para recorrer el inventario.

6. calcular_valor_inventario(self): Calcula y muestra el valor total del inventario.
  - Utiliza un bucle for para calcular el valor total del inventario.
  - Itera a través del inventario y suma el valor de cada producto (precio x cantidad). Es decir,calcula el valor total del inventario.

```python
class TiendaOnline: # creo la tienda online utilizando class.
    """ defino la función con el método constructor __init__  para definir los atributos que tendrá la tienda online.
    Que en este caso es inventario (solo he usado este en el ejercicio) y ventas_totales"""
    def __init__(self): 
        self.inventario = []
        self.ventas_totales = 0.0

    """Defino una funcion, incluyo dentro de los parentesis los atributos que voy a usar en la funcion.
        Añado un bucle for para recorrer la lista con diccionarios. (Aún no hay diccionarios)
        Utilizo la condición if para que cuando encuentre el articulo que he pedido le sume la cantidad que he introducido
        y si el articulo ya existe me lo imprime
        Terminamos con un else, se ejecuta cuando al recorrer la lista no encuentra el articulo, entonces lo añade con el metodo append. """
    def agregar_producto(self, nombre, precio, cantidad):
        for articulo in self.inventario:
            if articulo["nombre"] == nombre: # metodo para indicar la key del diccionario y te devuelve el value.
                articulo["cantidad"] += cantidad # sumamos a la cantidad del diccionario a la cantidad introducida.
                print(f"Este articulo ya existe, has añadido {cantidad} {nombre} al inventario")
                break # utilizamos break para que cuando encuentre el articulo se pare el bucle y salte a la siguiente acción.
        else:
            articulo_nuevo ={"nombre" : nombre, "precio" : precio, "cantidad" : cantidad} #creo un diccionario para introducir los elementos nuevos.
            self.inventario.append(articulo_nuevo) # metodo agregar para introducir los datos en el nuevo diccionario
            print(f"Has añadido {cantidad} {nombre.title()} a tu inventario") # .title() lo vi en el ejercicio y me parecio lo correcto.
            
    """definimos la función, no intruducimos los tributos en los parentesis porque quiero verlos todos.
        bucle for para que recorra mi inventario, articulo variable que puedo manipular.
        al acceder al bucle compara si los articulo introducidos son iguales que los que ya estan en el inventario
        cuando los encuentra los imprime, no hace falta break porque quiero que lea toda la informacion dentro de diccionario"""
    def ver_inventario(self):
        for articulo in self.inventario:
            nombre = articulo["nombre"]
            precio = articulo["precio"]
            cantidad = articulo["cantidad"]
            print(f"El inventario contiene: {nombre.title()}, Precio: {precio} € , Cantidad : {cantidad}")

    """Definimos la funcion, entre parectesis solo lo que vamos a buscar
        """
    def buscar_producto(self, nombre):
        
        for articulo in self.inventario:
            if articulo["nombre"].title() == nombre.title():
                print(f"El articulo buscado es: {articulo['nombre'].title()}, Precio: {articulo['precio']}€, Cantidad: {articulo['cantidad']}")
                break
                
            else:
                print(f"El articulo: {nombre} no esta en el inventario")


    def actualizar_stock(self, nombre, cantidad):
        articulo_inventario = False
        
        for articulo in self.inventario:
            if articulo["nombre"].title() == nombre.title():
                articulo_inventario = True

                if cantidad < 0:

                    if (int(articulo["cantidad"]) + int(cantidad)) < 0:
                    
                        print(f"No tienes suficiente stock. Hay en el inventario: {articulo["nombre"]} {articulo["cantidad"]}")

                    else:
                        articulo["cantidad"] += cantidad
                        print(f"Has actualizado el stock. Hay en el inventario:{articulo["nombre"]} {articulo["cantidad"]}")

                elif cantidad > 0:
                    articulo["cantidad"] += int(cantidad)
                    print(f"Has actualizado el stock. Hay en el inventario:{articulo["nombre"]} {articulo["cantidad"]}")
                else:
                    print(f"Has introducido 0 articulos. El inventario no ha cambiado")
                    break

        if not articulo_inventario:
            print(f"Articulo {nombre} no esta en el iventario")


    def eliminar_producto(self, nombre):
        articulo_inventario = False

        for articulo in self.inventario:
            if articulo["nombre"].title() == nombre.title():
                self.inventario.remove(articulo)
                print(f"Has eliminado el articulo ({articulo["nombre"]}) de tu inventario")
                articulo_inventario = True
                break

        if not articulo_inventario:
            print(f"No pudiste eliminar el articulo ({nombre}) ya que no esta en el inventario")


    def calcular_valor_inventario(self):
        total = 0
        for articulo in self.inventario:
            total_articulo = float(articulo["precio"]) * float(articulo["cantidad"])
            total += total_articulo

        print(f"El inventario tiene en total: {round(total, 2)}€")
        
```
