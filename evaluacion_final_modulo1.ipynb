Ejercicio
A lo largo de esta evaluación tendrás que crear una clase llamada TiendaOnline que cumpla los siguientes requisitos:
· La clase TiendaOnline debe tener los siguientes atributos:
1. inventario(lista de diccionarios): Un atributo para almacenar los productos en el inventario. Cada producto debe ser representado como un diccionario con las siguientes claves:'nombre', 'precio', y cantidad' Al principio deberá ser una lista vacía. Ejemplo decomo debería ser:

2. ventas_totales(float): Un atributo para llevar un registro de las ventas totales de la tienda.Inicializado con valor 0.

· La clase TiendaOnline debe tener los siguientes métodos:

1. agregar_producto(self, nombre, precio, cantidad): Este método agrega un producto alinventario o actualiza su cantidad si ya existe. Debe recibir el nombre, precio y cantidad del productocomo parámetros.
      -Itera a través del inventario y compara los nombres de los productos con el nombre proporcionado.
      -Si el producto ya existe, actualiza la cantidad.
      -Si no existe, agrega un nuevo producto al inventario.

2. ver_inventario(self): Muestra el inventario de productos con sus detalles.
· Utiliza un bucle for para recorrer el inventario.
· Imprime los detalles (nombre, precio, cantidad) de cada producto.

3.  buscar_producto(self, nombre): Busca un producto en el inventario por nombre y muestra susdetalles si se encuentra. Debe recibir el nombre del producto como parámetro.
   - Utiliza un bucle for para recorrer el inventario.
   - Compara los nombres de los productos con el nombre proporcionado.
   - Si se encuentra el producto, imprime sus detalles.

4. actualizar_stock(self, nombre, cantidad): Actualiza el stock de un producto en elinventario. Debe recibir el nombre del producto y la cantidad a agregar o quitar como parámetros.
  - Utiliza un bucle for para recorrer el inventario.
  - Busca el producto por nombre.
  - Actualiza la cantidad según la entrada del usuario.
  - Si el producto no esta en el inventario muestra un mensaje indicándolo.

5. eliminar_producto(self, nombre): Elimina un producto del inventario por nombre. Debe recibir el nombre del producto como parámetro.
  - Utiliza un bucle for para recorrer el inventario.
  - Busca el producto por nombre.
  - Elimina el producto del inventario si existe.
  - Si el producto no esta en el inventario muestra un mensaje indicándolo.

6. calcular_valor_inventario(self): Calcula y muestra el valor total del inventario.
  - Utiliza un bucle for para calcular el valor total del inventario.
  - Itera a través del inventario y suma el valor de cada producto (precio x cantidad). Es decir,calcula el valor total del inventario.

```python
class TiendaOnline: # creo la tienda online utilizando class.
    # defino la función con el método constructor __init__  para definir los atributos que tendrá la tienda online.
    def __init__(self): 
        self.inventario = []
        self.ventas_totales = 0.0

    """ definimos la funcion con los tres atributos para agregarlos a un diccionario dentro de una lista
        bucle for para recorrer la lista
        condicional if para que encuentra el articulo le añada la cantidad que hemos proporcionado
        break para parar el bucle cuando la encuentre
        else para añadir si el producto no existe, creamos un diccionario con las claves : valor de nuestros atributos
        metodo append para agregar el articulo."""

    def agregar_producto(self, nombre, precio, cantidad):
        for articulo in self.inventario:

            if articulo["nombre"] == nombre:
                articulo["cantidad"] += cantidad
                print(f"Este articulo ya existe, has añadido {cantidad} {nombre} al inventario")
                break

        else:
            articulo_nuevo ={"nombre" : nombre, "precio" : precio, "cantidad" : cantidad}
            self.inventario.append(articulo_nuevo)
            print(f"Has añadido {cantidad} {nombre.title()} a tu inventario")
            
    """ definimos la funcion, sin atributos porque queremos que nos muestre todo lo que hay en inventario
        bucle for para que recorra la lista 
        variables para que muestre el value de mi diccionario."""
    def ver_inventario(self):

        for articulo in self.inventario:
            nombre = articulo["nombre"]
            precio = articulo["precio"]
            cantidad = articulo["cantidad"]
            print(f"El inventario contiene: {nombre.title()}, Precio: {precio} € , Cantidad : {cantidad}")

    """definimos la funcion, solo el atributo nombre para buscar
        creamos un bucle for para buscar en la lista 
        if condicion si el nombre introducido es igual al que esta en la lista
        else si el articulo no esta
        retur para salir del bucle cuando haya encontrado el articulo"""
    
    def buscar_producto(self, nombre):
        
        for articulo in self.inventario:
            if articulo["nombre"].title() == nombre.title():
                print(f"El articulo buscado es: {articulo['nombre'].title()}, Precio: {articulo['precio']}€, Cantidad: {articulo['cantidad']}")
                return
                
            else:
                print(f"El articulo: {nombre.title()} no esta en el inventario")
                return
            
    """Definimos la funcion con los atributos que queremos actualizar
        creamos bucle for para recorrer la lista
        condicion if para si el nombre coincide, otra condicion si coincide el articulo
         if para si al sumar la cantidad que hay y la cantidad que añadiños es menor a cero no se pueda realizar la accion
         elif por si introducimos 0
         else para añadir los articulos
         return para salir del bucle y si no esta pues imprime el mensaje"""   
        
    def actualizar_stock(self, nombre, cantidad):
        
        for articulo in self.inventario:
            if articulo["nombre"].title() == nombre.title():
                
                if articulo["cantidad"] + cantidad < 0:

                    print(f"No tienes suficiente stock. Hay en el inventario: Articulo: {articulo["nombre"].title()} / Cantidad :{articulo["cantidad"]}")

                elif cantidad == 0:
                     print(f"Has introducido 0 articulos. El inventario no ha cambiado")
                     return    

                else:
                    articulo["cantidad"] += cantidad
                    
                    print(f"Has actualizado el stock. Hay en el inventario: Articulo: {articulo["nombre"].title()} / Cantidad: {articulo["cantidad"]}")

                    return
        
        print(f"Articulo {nombre} no esta en el iventario")

    """definimos la funcion con el atributo nombre
        bucle for para recorrer la lista
        if condicinal para cuando encuentre el articulo dentro de la lista si es el mismo que hemos introducido lo elimine con el metodo .remove()
        return para salir del bucle y si no esta el articulo te devuelva el mensaje."""
    
    def eliminar_producto(self, nombre):

            for articulo in self.inventario:

                if articulo["nombre"].title() == nombre.title():
                    self.inventario.remove(articulo)
                    print(f"Has eliminado el articulo ({articulo["nombre"].title()}) de tu inventario")
                    return
                    
            print(f"No pudiste eliminar el articulo ({nombre}) ya que no esta en el inventario")

    """definimos la funcion, no hace falta atributos
        variable total a cero la usamos como contador
        bucle for para recorrer la lista
        variable total_articulo para definir la multiplicacion
        float para pasar los numeros a decimales y asi al hacer operaciones son todos iguales
        sumamos el total a el total de articulos de la lista y asi lo imprime, round para que devuelva solo dos decimales.
        """
    def calcular_valor_inventario(self):
        total = 0

        for articulo in self.inventario:
            total_articulo = float(articulo["precio"]) * float(articulo["cantidad"])
            total += total_articulo

        print(f"El inventario tiene en total: {round(total, 2)}€")
        
```
